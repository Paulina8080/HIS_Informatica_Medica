```python
def sistema_informacion():
    # Esta variable controla si el bucle sigue o no
    continuar = True
    
    while continuar:
        print("\n" + "="*30)
        print(" SISTEMA DE GESTIÓN (HIS) ")
        print("="*30)
        print("1. Ingresar información")
        print("2. Buscar información")
        print("3. Actualizar información")
        print("4. Eliminar información")
        print("5. Salir y cerrar celda")
        
        opcion = input("Seleccione una opción (1-5): ")

        if opcion == "1":
            print("\n[MÓDULO] Ingresando nuevos datos...")
        elif opcion == "2":
            print("\n[MÓDULO] Buscando en la base de datos...")
            print("Esto es una prueba para verificar conexión con Github")
        elif opcion == "3":
            print("\n[MÓDULO] Actualizando registros...")
        elif opcion == "4":
            print("\n[MÓDULO] Eliminando información...")
        elif opcion == "5":
            print("\nCerrando sesión... La celda terminará de ejecutarse ahora.")
            continuar = False  # Esto rompe el bucle y detiene la celda
        else:
            print("\n[!] Opción no válida. Intente de nuevo.")

# Llamamos a la función
sistema_informacion()
```
