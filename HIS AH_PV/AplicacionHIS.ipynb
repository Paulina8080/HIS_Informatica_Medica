```python
try:
    while True:
        print("\n" + "="*35)
        print("   SISTEMA DE GESTIÓN (HIS)   ")
        print("="*35)
        print("1. Ingresar información")
        print("2. Buscar información")
        print("3. Actualizar información")
        print("4. Eliminar información")
        print("5. Salir")
        
        # RECUERDA: Escribe el número en la barra que sale ARRIBA en VS Code
        opcion = input("Seleccione una opción (1-5): ")

        if opcion == "1":
            print("\n[OK] Abriendo módulo de ingreso...")
        elif opcion == "2":
            print("\n[OK] Buscando en la base de datos...")
        elif opcion == "3":
            print("\n[OK] Editando registros...")
        elif opcion == "4":
            print("\n[OK] Registro eliminado.")
        elif opcion == "5":
            print("\nCerrando sistema... Sesión finalizada.")
            break
        else:
            print("\n[!] Opción no válida, intenta de nuevo.")
except KeyboardInterrupt:
    print("\n\nProceso detenido manualmente.")
```
